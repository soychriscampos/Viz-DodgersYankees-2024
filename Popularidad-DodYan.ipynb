# Popularidad de Dodgers vs Yankees en México

El objetivo de este notebook es crear una visualizaicón con matplotlib de la popularidad de ambos equipos desde que inició la liga hasta el día del juego 1 de la Serie Mundial 2024. El dataset ha sido obtenido de Google trends.

## Los Datos

Librerías necesarías:

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

Los datos prácticamente están listos para trabajar, lo único que realicé fue saltar un tipo de etiqueta que arroja google trends por defecto, renombrar algunas columnas, y modificar el tipo de dato de fecha para poder manipularlo. Ésto último resultó muy útil ya que los datos descargados habían sido desde del año 2023, que para los propósitos de la visualización no eran relevantes.

```python
df = pd.read_csv('timeline.csv', skiprows=1)

# Limpieza
df.columns = df.columns.str.strip()
df = df.rename(columns={'Dodgers: (México)' : 'Dodgers', 'Yankees: (México)': 'Yankees'})
df['Semana'] = pd.to_datetime(df['Semana'])

# Filtrado
df = df[df['Semana'] >= '2024-03-01']
```

```python
plt.figure(figsize=(10 , 6))

# Fondo
plt.gca().set_facecolor('#e1d8be')

# Quitar Marcos
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_visible(False)

# Gráfico de datos
plt.plot(df['Semana'], df['Dodgers'], label='Dodgers', color='#005a9c', linewidth=2)
plt.plot(df['Semana'], df['Yankees'], label='Yankees', color='#e4002b', linewidth=2)

# Título, etiquetas y leyenda
plt.title('Popularidad de Dodgers vs Yankees en México\n', fontsize = 16, color='#08144d', fontweight='bold')
plt.text(0.5, 1.02, 'Desde que inició la liga hasta el día del juego #1 de la Serie Mundial 2024.',
        fontsize=12, color='#08144d', ha='center', va='bottom', transform=ax.transAxes)
plt.ylabel('Interés de búsqueda (%)', fontsize = 10, color='#08144d')
plt.legend(loc='upper left', frameon=False, facecolor='#c8b698', edgecolor='#c8b698', labelcolor = '#08144d')

# Cuadrícula y fechas
plt.grid(True, linestyle='--', alpha=0.5)
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b-%Y'))
plt.xticks(rotation=10, fontsize=10, color = '#08144d')

# Logos
logo_dodgers = mpimg.imread('dodgers_big.png')
logo_yankees = mpimg.imread('yankees_big.png')

imagebox_dodgers = ax.inset_axes([0.4,0.5,0.1,0.5])
imagebox_dodgers.imshow(logo_dodgers)
imagebox_dodgers.axis('off')

imagebox_yankees = ax.inset_axes([0.5,0.5,0.1,0.5])
imagebox_yankees.imshow(logo_yankees)
imagebox_yankees.axis('off')

# Fuente de datos y firma
plt.text(0.5, -0.1, '\n\n\nFuente: Google Trends.\nElaborado por: Christian Campos.',
        fontsize=6, color='#08144d', ha='center', va='center', transform=ax.transAxes)

plt.tight_layout()

plt.savefig('popularidad_dodgers_vs_yankees.png', dpi=300)

plt.show()
```

## Conclusión
Aunque ambos equipos presentan picos de interés a lo largo de la temporada, los Dodgers superan consistentemente a los Yankees en popularidad en la mayoría de los meses.
